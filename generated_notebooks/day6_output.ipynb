# Paper Summary

- Formulates bbFMM as an \(N\)-body summation problem \(f(x_i)=\sum_{j=1}^N K(x_i,y_j)\sigma_j\) and replaces the naïve \(O(N^2)\) matrix–vector product with a hierarchical, interpolation-based scheme that achieves \(O(N)\) complexity for fixed accuracy.  

- Uses Chebyshev nodes and the truncated Chebyshev reproducing kernel \(S_n\) to build a numerical low-rank approximation of the kernel on well-separated source–target intervals: in 1D \(K(x,y)\approx\sum_{l,m=1}^n K(x_l,y_m)S_n(x_l,x)S_n(y_m,y)\), and in \(d\) dimensions via tensor products with \(n^d\) nodes per box.  

- Represents each well-separated box pair \((I,J)\) by an \(n^d\times n^d\) M2L sampling matrix \(K^{(I,J)}_{lm}=K(x^{(I)}_l,y^{(J)}_m)\), and defines multipole coefficients \(W^{(J)}_m\) (P2M) and local coefficients \(g^{(I)}_l\) (M2L) as algebraic objects living on Chebyshev grids rather than analytic series.  

- Organizes computation on a tree of boxes (quadtree/octree in practice), classifying interactions into near-field (direct particle–particle or node–node) and far-field (interpolated) via a geometric well-separatedness criterion based on box distance versus diameter, so that interpolation is only used where the kernel is smooth.  

- Defines a three-pass multilevel algorithm:  
  - Upward pass: P2M anterpolates particles to leaf-box Chebyshev nodes, then M2M translates multipoles from children to parents using interpolation matrices \(A^{(P,c)}\).  
  - Interaction pass: for each box, accumulates far-field contributions from its interaction list by applying the precomputed M2L matrices to source multipoles.
