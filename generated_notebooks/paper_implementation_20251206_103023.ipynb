## 📄 Paper Summary

- **Chebyshev-based kernel compression:** Far-field interactions are approximated via tensor-product Chebyshev interpolation of the kernel \(K(x,y)\), using the truncated reproducing kernel \(S_n\) in 1D and its tensorized form \(R_n\) in \(d\) dimensions, yielding a numerical low-rank factorization
  \[
  K(x,y) \approx \sum_{l,m=1}^n K(x_l,y_m) S_n(x_l,x) S_n(y_m,y).
  \]

- **Hierarchical tree and interaction classification:** The domain is recursively partitioned into a \(2^d\)-ary tree; interactions are split into **near-field** (direct summation over neighbors) and **far-field** (Chebyshev-compressed) using a geometric well-separatedness condition \( \text{dist}(S,T) \ge \varepsilon \min\{\text{diam}(S),\text{diam}(T)\}\).

- **Algebraic operators on Chebyshev nodes:** Each box stores \(n^d\) Chebyshev nodes; the algorithm defines:
  - **P2M:** anterpolation \(W_m^{(J)} = \sum_{y_j\in J} \sigma_j S_n(y_m^{(J)}, y_j)\),
  - **M2M:** parent aggregation via translation matrices \(A_{m,m'}^{(P,c)} = S_n(y_m^{(P)}, y_{m'}^{(c)})\),
  - **M2L:** box-pair interaction with precomputed sampling matrices \(K^{(I,J)}_{lm} = K(x_l^{(I)}, y_m^{(J)})\),
  - **L2L:** downward translation via \(B_{l,l'}^{(c,P)} = S_n(x_l^{(c)}, x_{l'}^{(P)})\),
  - **L2P:** interpolation at targets using \(S_n(x_l^{(I)}, x_i)\) plus direct near-field corrections.

- **Three-pass multilevel algorithm:**  
  1) **Upward pass:** P2M at leaves then M2M to propagate multipole coefficients to the root.  
  2) **Interaction pass:** for each box, accumulate far-field contributions from its interaction
